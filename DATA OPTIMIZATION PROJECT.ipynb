```python
BREAK_DATE = '2026-01-10'
break_date = pd.to_datetime(BREAK_DATE)
from scipy.stats import mannwhitneyu

# Split into pre‑ and post‑break
pre = df[df['usage_date'] < break_date].copy()
post = df[df['usage_date'] >= break_date].copy()

# Add day-of-week names
pre['weekday'] = pre['usage_date'].dt.day_name()
post['weekday'] = post['usage_date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
             'Friday', 'Saturday', 'Sunday']

print("=" * 70)
print("Stability of Day‑of‑Week Pattern Across Regimes")
print("=" * 70)
print(f"Pre‑break: {len(pre)} days ({pre['usage_date'].min().date()} to {pre['usage_date'].max().date()})")
print(f"Post‑break: {len(post)} days ({post['usage_date'].min().date()} to {post['usage_date'].max().date()})")
print()

# ------------------------------------------------------------------
# 2. Compute medians and 90th percentiles per weekday
# ------------------------------------------------------------------
pre_stats = pre.groupby('weekday')['usage_mb'].agg(['median', lambda x: x.quantile(0.9)]).rename(columns={'median': 'median_pre', '<lambda>': 'p90_pre'})
post_stats = post.groupby('weekday')['usage_mb'].agg(['median', lambda x: x.quantile(0.9)]).rename(columns={'median': 'median_post', '<lambda>': 'p90_post'})

stats = pd.concat([pre_stats, post_stats], axis=1).reindex(day_order)
print("Day-of-week medians and 90th percentiles (MB):")
print(stats.round(1))

# ------------------------------------------------------------------
# 3. Mann‑Whitney U tests for each weekday
# ------------------------------------------------------------------
print("\n" + "=" * 70)
print("Mann‑Whitney U tests (p‑value < 0.05 indicates distribution change)")
print("=" * 70)

results = []
for day in day_order:
    pre_vals = pre[pre['weekday'] == day]['usage_mb'].values
    post_vals = post[post['weekday'] == day]['usage_mb'].values
    if len(pre_vals) > 0 and len(post_vals) > 0:
        stat, p = mannwhitneyu(pre_vals, post_vals, alternative='two-sided')
        results.append({'weekday': day, 'p-value': p, 'n_pre': len(pre_vals), 'n_post': len(post_vals)})
    else:
        results.append({'weekday': day, 'p-value': np.nan, 'n_pre': len(pre_vals), 'n_post': len(post_vals)})

df_results = pd.DataFrame(results).set_index('weekday').reindex(day_order)
print(df_results.round(4))

# ------------------------------------------------------------------
# 4. Visual comparison (optional)
# ------------------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Boxplots for each period separately
ax = axes[0]
sns.boxplot(data=pre, x='weekday', y='usage_mb', order=day_order, ax=ax, palette='Blues')
ax.set_title('Pre‑break Usage by Day')
ax.set_ylabel('Usage (MB)')
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=45)

ax = axes[1]
sns.boxplot(data=post, x='weekday', y='usage_mb', order=day_order, ax=ax, palette='Reds')
ax.set_title('Post‑break Usage by Day')
ax.set_ylabel('Usage (MB)')
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# ------------------------------------------------------------------
# 5. Interpretation
# ------------------------------------------------------------------
print("\n" + "=" * 70)
print("Interpretation")
print("=" * 70)
significant = df_results[df_results['p-value'] < 0.05].index.tolist()
if len(significant) > 0:
    print(f"Significant differences (p < 0.05) found for: {', '.join(significant)}")
    print("Therefore we will keep pre‑ and post‑break periods separate for distributional testing.")
else:
    print("No significant differences detected. We may pool pre‑ and post‑break data for lognormal testing.")
```

```python
"""
Step 2 – Statistical Synthesis: Test lognormal distribution per weekday (post‑break only)
--------------------------------------------------------------------------------------------
For each weekday in the post‑break period:
- Take the usage values (MB).
- Transform by natural log.
- Visual: Q‑Q plot against a normal distribution.
- Formal: Shapiro‑Wilk test for normality of the log values.
- Because sample sizes are small, we rely primarily on the Q‑Q plots.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
from scipy.stats import shapiro

# ------------------------------------------------------------------
# 1. Prepare post‑break data (assuming df_full exists)
# ------------------------------------------------------------------
# df_full = ... # your full usage DataFrame with columns 'usage_date', 'usage_mb'
# Ensure sorted
df = df.sort_values('usage_date').reset_index(drop=True)

BREAK_DATE = '2026-01-10'
break_date = pd.to_datetime(BREAK_DATE)

post = df[df['usage_date'] >= break_date].copy()
post['weekday'] = post['usage_date'].dt.day_name()
post['log_usage'] = np.log(post['usage_mb'])

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
             'Friday', 'Saturday', 'Sunday']

print("=" * 70)
print("Testing Lognormal Assumption per Weekday (Post‑Break Only)")
print("=" * 70)
print(f"Post‑break period: {post['usage_date'].min().date()} to {post['usage_date'].max().date()}")
print(f"Total days: {len(post)}")

# ------------------------------------------------------------------
# 2. Shapiro‑Wilk test for each weekday
# ------------------------------------------------------------------
results = []
fig, axes = plt.subplots(4, 2, figsize=(12, 12)) # 4x2 grid for Q‑Q plots
axes = axes.flatten()

for i, day in enumerate(day_order):
    data = post[post['weekday'] == day]['log_usage'].dropna()
    n = len(data)
    if n >= 3: # Shapiro‑Wilk requires at least 3 observations
        stat, p = shapiro(data)
        results.append({'weekday': day, 'n': n, 'statistic': stat, 'p‑value': p})
    else:
        results.append({'weekday': day, 'n': n, 'statistic': np.nan, 'p‑value': np.nan})
        p = np.nan

    # Q‑Q plot
    ax = axes[i]
    stats.probplot(data, dist="norm", plot=ax)
    ax.set_title(f'{day} (n={n})\nShapiro‑Wilk p={p:.3f}' if not np.isnan(p) else f'{day} (n={n})\nInsufficient data')
    ax.grid(True, alpha=0.3)

# Hide any unused subplots (if any)
for j in range(len(day_order), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()

# ------------------------------------------------------------------
# 3. Display test results table
# ------------------------------------------------------------------
df_results = pd.DataFrame(results).set_index('weekday').reindex(day_order)
print("\nShapiro‑Wilk Test Results (for log‑transformed usage):")
print(df_results.round(4))

# ------------------------------------------------------------------
# 4. Interpretation
# ------------------------------------------------------------------
print("\n" + "=" * 70)
print("Interpretation")
print("=" * 70)
print("The Q‑Q plots show how closely the log‑transformed usage follows a normal distribution.")
print("Points that lie close to the diagonal support the lognormal assumption.")
print("The Shapiro‑Wilk p‑values are reported but have low power due to small sample sizes.")
print("If the plots appear reasonably linear, we can consider the lognormal assumption as a plausible description.")
```

```python
#Step 3 – Statistical Synthesis: Bootstrap Simulation of Monthly Totals

#Generates an empirical distribution of monthly data usage by resampling
#the post‑break daily values. No distributional assumptions are needed;
#the output will be used directly in the optimisation stage.


BREAK_DATE = '2026-01-10'
break_date = pd.to_datetime(BREAK_DATE)

post = df[df['usage_date'] >= break_date].copy()
daily_vals = post['usage_mb'].values
print(f"Post‑break period: {post['usage_date'].min().date()} to {post['usage_date'].max().date()}")
print(f"Number of days: {len(daily_vals)}")


# 2. Bootstrap simulation

n_sim = 10000
days_in_month = 30
monthly_totals = np.zeros(n_sim)

for i in range(n_sim):
    # Sample 30 days with replacement from the post‑break daily values
    sample = np.random.choice(daily_vals, size=days_in_month, replace=True)
    monthly_totals[i] = sample.sum()

monthly_gb = monthly_totals / 1000.0

# 3. Summary statistics
mean_gb = monthly_gb.mean()
median_gb = np.median(monthly_gb)
std_gb = monthly_gb.std()
p95_gb = np.percentile(monthly_gb, 95)

print("\n" + "=" * 70)
print("Bootstrap Results (10,000 simulated months)")
print("=" * 70)
print(f"Mean monthly need: {mean_gb:.2f} GB")
print(f"Median monthly need: {median_gb:.2f} GB")
print(f"Std deviation: {std_gb:.2f} GB")
print(f"95th percentile: {p95_gb:.2f} GB ← target for conservative planning")

# ------------------------------------------------------------------
# 4. Visualisation
# ------------------------------------------------------------------
plt.figure(figsize=(10, 5))
plt.hist(monthly_gb, bins=50, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(p95_gb, color='red', linestyle='--', linewidth=2, label=f'95th percentile: {p95_gb:.1f} GB')
plt.xlabel('Monthly data usage (GB)')
plt.ylabel('Density')
plt.title('Bootstrap Distribution of Monthly Totals (Post‑Break)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#"""
#Data Portfolio Optimizer – Final Blended Effective Cost (BEC) Optimisation
#----------------------------------------------------------------------------
#Uses bootstrap monthly totals to simulate months, and for months where the
#base plan runs out, determines the cheapest valid top‑up plan based on:
#    - Remaining days in the month
 #   - Required additional GB

#The top‑up plan is chosen from the available plans (daily, 2‑3 day, weekly)
#that can cover the remaining days and the shortfall volume.

#Blended Effective Cost (BEC) = (Σ (P_base + P_top)) / (Σ U_i)
#where P_top = 0 for months where U_i ≤ V_base.
#"""

import numpy as np
import pandas as pd
from itertools import product

# ------------------------------------------------------------------
# 1. Bootstrap monthly totals (replace with your actual array)
# ------------------------------------------------------------------
# Simulated for demonstration – replace with your real bootstrap array


# ------------------------------------------------------------------
# 2. Define all available plans (volume in GB, price in ₦, validity in days)
# ------------------------------------------------------------------
plans_db = [
    {'name': 'Daily 1GB', 'volume': 1.0, 'price': 500, 'validity': 1},
    {'name': 'Daily 3.5GB', 'volume': 3.5, 'price': 1000, 'validity': 1},
    {'name': '2‑3d 7GB', 'volume': 7.0, 'price': 1800, 'validity': 3}, # assume 3 days for simplicity
    {'name': 'Weekly 20GB', 'volume': 20.0, 'price': 5000, 'validity': 7},
    # Monthly plans are the base plans we test
]

# Base plans to test (these are the ones we consider buying at the start of the month)
base_plans = [
    {'name': 'Monthly 14.5GB', 'volume': 14.5, 'price': 5000},
    {'name': 'Monthly 24GB', 'volume': 24.0, 'price': 7500},
    {'name': 'Monthly 65GB', 'volume': 65.0, 'price': 16000},
    {'name': 'Monthly 165GB', 'volume': 165.0, 'price': 35000},
]

# Combo strategies (pre‑bought base + booster) – we treat these as single base plans
combos = [
    {'name': '14.5GB + 1×7GB', 'volume': 14.5 + 7.0, 'price': 5000 + 1800},
    {'name': '14.5GB + 2×7GB', 'volume': 14.5 + 14.0, 'price': 5000 + 3600},
    {'name': '4×7GB (no base)', 'volume': 28.0, 'price': 4 * 1800},
]

all_base = base_plans + combos

# ------------------------------------------------------------------
# 3. Function to find cheapest top‑up plan given remaining days and required GB
# ------------------------------------------------------------------
def cheapest_topup(required_gb, remaining_days):
    """
    Returns the cheapest plan (price, name) that covers at least required_gb
    and has validity ≥ remaining_days.
    """
    candidates = []
    for plan in plans_db:
        if plan['volume'] >= required_gb and plan['validity'] >= remaining_days:
            candidates.append((plan['price'], plan['name']))
    if not candidates:
        # If no single plan covers both, we may need to stack – for simplicity,
        # we fall back to the cheapest per‑GB plan that covers the volume.
        # But in practice, with our plans, this shouldn't happen for realistic shortfalls.
        # We'll use the cheapest daily per GB.
        daily = [p for p in plans_db if p['validity'] >= remaining_days]
        if daily:
            best = min(daily, key=lambda x: x['price'] / x['volume'])
            return best['price'], best['name']
        else:
            # Absolute fallback: use the smallest daily
            smallest = min(plans_db, key=lambda x: x['price'])
            return smallest['price'], smallest['name']
    return min(candidates, key=lambda x: x[0])

# ------------------------------------------------------------------
# 4. Day‑by‑day simulation to find when shortfall occurs
# ------------------------------------------------------------------
def simulate_month_for_base(base_plan, monthly_gb, days_in_month=30):
    """
    Simulates one month with the given base plan.
    Returns (total_cost, total_usage) for that month.
    """
    V = base_plan['volume']
    P = base_plan['price']
    
    # Total usage for the month (from bootstrap)
    total_usage = monthly_gb
    
    if total_usage <= V:
        # No shortfall
        return P, total_usage
    
    # Shortfall occurred – we need to find when during the month
    # We'll use the daily usage pattern to approximate the timing.
    # Since we don't have daily sequence for this specific month, we use the average
    # daily usage distribution (from post‑break data) to estimate the day when we run out.
    
    # For simplicity, we assume usage is uniform across the month. This is an approximation,
    # but it's acceptable because the exact day of shortfall has limited impact on top‑up
    # plan choice – the key is remaining days.
    
    daily_avg = total_usage / days_in_month
    days_until_exhaustion = int(V / daily_avg) # floor
    remaining_days = days_in_month - days_until_exhaustion
    
    required_gb = total_usage - V
    
    topup_price, _ = cheapest_topup(required_gb, remaining_days)
    
    total_cost = P + topup_price
    return total_cost, total_usage

# ------------------------------------------------------------------
# 5. Run BEC calculation for each base plan
# ------------------------------------------------------------------
results = []
n_months = len(monthly_gb)

for base in all_base:
    total_cost = 0
    total_usage = 0
    for usage in monthly_gb:
        cost, used = simulate_month_for_base(base, usage)
        total_cost += cost
        total_usage += used
    
    bec = total_cost / total_usage
    results.append({
        'Plan': base['name'],
        'Volume (GB)': base['volume'],
        'Price (₦)': base['price'],
        'BEC (₦/GB)': bec,
        'Total Cost (₦)': total_cost,
        'Total Usage (GB)': total_usage
    })

df_results = pd.DataFrame(results)
df_results = df_results.sort_values('BEC (₦/GB)')

# ------------------------------------------------------------------
# 6. Output results
# ------------------------------------------------------------------
print("=" * 80)
print("BLENDED EFFECTIVE COST (BEC) OPTIMISATION")
print("=" * 80)
print(df_results.round(2).to_string(index=False))

print("\n" + "=" * 80)
print("RECOMMENDATION")
print("=" * 80)
best = df_results.iloc[0]
print(f"The plan with the lowest Blended Effective Cost is:")
print(f" → {best['Plan']}")
print(f" Volume: {best['Volume (GB)']:.1f} GB")
print(f" Price: ₦{best['Price (₦)']:,.0f}")
print(f" BEC: ₦{best['BEC (₦/GB)']:.2f} per used GB")
print(f" Total cost over {n_months} months: ₦{best['Total Cost (₦)']:,.0f}")
print(f" Total usage over {n_months} months: {best['Total Usage (GB)']:.0f} GB")
```
